```python
months = ['сентября', 'октября', 'ноября', 'декабря', 'января', 'февраля', 'марта', 'апреля', 'мая', 'июня', 'июля', 'августа']
months
```

```python
tmp_df = []
for col_num in df.columns:
    dates = []
    selected_rows = []
    rows_num = len(df.index) + 1
    for row_num, row in df.iterrows():
        cell_val = row[col_num]
        try:
            if cell_val.split(' ')[1] in months:
                dates.append(cell_val)
                selected_rows.append(row_num + 1)
        except:
            pass
    selected_rows.append(rows_num)
    tmp_df += [(dates[i], df[col_num][selected_rows[i]:selected_rows[i+1] - 1]) for i in range(len(dates))]
        
```
